# Solution to the Birthdays Problem

```python
import numpy as np
import matplotlib.pyplot as plt
```

The logic:

* People are showing up one by one (we pretend that they are showing up one by one, as it makes thinking easier)
* Once i-th person comes into the room, there are i-1 people already there. 
* Let's calculate P(that there's no overlap in birthdays). Each person "claims" a day of the year.
* With i-1 people, it means i-1 days taken. So (i-1)/365 share of the year is taken.
* The P of a new person not falling into the "taken" part of the year is therefore 1-(i-1)/365
* So we have enough to calculate it. Each new (ith) person maintains the status quo of "no overlap in BDs" with 1-(i-1)/365 probability. And originally, for i=1, P(no overlap) = 1

```python
maxn = 80              # Highest number of ppl we'll consider
p = np.zeros(maxn)      # Future answer, p(n). We create an array and fill it with zeros
p[0] = 1                # If no people, p=1
p[1] = 1                # One person => also no overlap
for i in range(2,maxn):
    p[i] = p[i-1]*(1-(i-1)/365)
    
# The original question was about P(overlap), not P(no overlap)
plt.plot(1-p);
plt.xlabel('N people in the group');
plt.ylabel('P of no shared BD');
```
